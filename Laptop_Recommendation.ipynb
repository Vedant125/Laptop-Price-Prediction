```python
def cpu(x):
    if 'Intel Core i5' in x or 'Intel Core i3' in x or "Intel Core i7" in x or 'Intel Core i9' in x : 
        return x[0:13]
    elif "Intel" in x:
        return "Other Intel Processor"
    elif "AMD Ryzen 3" in x or 'AMD Ryzen 5' in x or 'AMD Ryzen 7' in x or 'AMD Ryzen 9' in x:
        return x[0:11]
    elif 'AMD' in x : 
        return 'Other AMD Processor'
    else:
        return "Other"
```

```python
df['Processor_Type'] = df['processor'].apply(cpu)
```

```python


def SSD(x):
    if 'SSD' in x:
        if '|' not in x:
            idx = x.index('SSD')
            return x[0:idx].strip()
        elif '|' in x:
            parts = x.split("|")
            for part in parts:
                if 'SSD' in part:
                    idx = part.index('SSD')
                    return part[0:idx].strip()
    return 0
```

```python
df['SSD(GB)'] = df['storage'].apply(lambda x : SSD(x))
```

```python
def HDD(x):
    if 'HDD' in x:
        if '|' not in x:
            idx = x.index('HDD')
            return x[0:idx].strip()
        elif '|' in x:
            parts = x.split("|")
            for part in parts:
                if 'HDD' in part:
                    idx = part.index('HDD')
                    return part[0:idx].strip()
    return 0
```

```python
df['HDD(GB)'] = df['storage'].apply(lambda x : HDD(x))
```

# Converting TB to GB

```python
def TB_TO_GB(x):
    x = str(x)
    if 'TB' in x : 
        return int(x.replace('TB' , '').strip()) * 1000
    else : 
        return int(x.replace('GB', '').strip())
        
```

```python
df['SSD(GB)'] = df['SSD(GB)'].apply(TB_TO_GB)
df['HDD(GB)'] = df['HDD(GB)'].apply(TB_TO_GB)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder

# # Load the dataset from CSV
# df = pd.read_csv('laptops.csv')

# One-hot encoding for categorical features
categorical_features = ['Processor_Type', 'OS', 'name']
one_hot_encoder = OneHotEncoder()
one_hot_encoded = one_hot_encoder.fit_transform(df[categorical_features]).toarray()

# Numerical features
numerical_features = df[['Ram(GB)', 'SSD(GB)', 'HDD(GB)']].values

# Combine encoded categorical features and numerical features
X = pd.concat([pd.DataFrame(one_hot_encoded), pd.DataFrame(numerical_features)], axis=1)

def recommend_laptops(user_input_list):
    # Convert user input into the same format as the dataset
    user_data = {
        'Processor_Type': user_input_list[1],
        'OS': user_input_list[3],
        'name': user_input_list[0],
        'Ram(GB)': user_input_list[2],
        'SSD(GB)': user_input_list[4],
        'HDD(GB)': user_input_list[5]
    }
    user_df = pd.DataFrame([user_data])
    user_one_hot_encoded = one_hot_encoder.transform(user_df[categorical_features]).toarray()
    user_numerical_features = user_df[['Ram(GB)', 'SSD(GB)', 'HDD(GB)']].values
    user_X = pd.concat([pd.DataFrame(user_one_hot_encoded), pd.DataFrame(user_numerical_features)], axis=1)
    
    # Calculate cosine similarity
    similarity_scores = cosine_similarity(user_X, X).flatten()
    
    # Add similarity scores to the dataframe
    df['similarity'] = similarity_scores
    
    # Sort by similarity and select top laptops from different companies
    top_laptops = df.sort_values(by='similarity', ascending=False).drop_duplicates(subset=['name']).head(5)
    
    # Return the recommendations with selected columns
    return top_laptops[['img_link', 'name', 'price(in Rs.)', 'Processor_Type', 'Ram(GB)', 'OS', 'SSD(GB)', 'HDD(GB)']]

# User input list
user_input_list = ['ASUS', 'Intel Core i7', 16, 'Windows', 256, 1000]

# Get recommendations
recommendations = recommend_laptops(user_input_list)
print(recommendations)
```
